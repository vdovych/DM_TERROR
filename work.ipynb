Transform npreps into processable columns with onehot encoding (npreps_unk,npreps_single,npreps_duo,npreps_small,npreps_medium,npreps_large)

```python
df_new["nperps_unk"] = df_new["nperps"] == -99
```

```python
df_new["nperps_single"] = df_new["nperps"] == 1
df_new["nperps_duo"] = df_new["nperps"] == 2
df_new["nperps_small"] = df_new["nperps"] < 5
df_new["nperps_medium"] = df_new["nperps"] < 30
df_new["nperps_large"] = df_new["nperps"] >= 30
```

```python
df_new["nperps_medium"] = df_new["nperps_medium"] != df_new["nperps_small"]
```

```python
df_new["nperps_small"] = df_new["nperps_small"] != df_new["nperps_single"]
df_new["nperps_small"] = df_new["nperps_small"] != df_new["nperps_duo"]
df_new["nperps_small"] = df_new["nperps_small"] != df_new["nperps_unk"]
```
